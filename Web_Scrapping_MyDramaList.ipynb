## Importing Libraries


```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

## Headers for the requests

```python
headers = {'user-agent':'Mozilla/5.0 \
            (Windows NT 10.0; Win64; x64) \
            AppleWebKit/537.36 (KHTML, like Gecko) \
            Chrome/84.0.4147.105 Safari/537.36'}
```

## List of URLs

```python
urls = [
    'https://mydramalist.com/62623-law-school',
    'https://mydramalist.com/33768-july-found-by-chance',
    'https://mydramalist.com/33645-a-little-thing-called-first-love',
    'https://mydramalist.com/696659-two-people-s-little-forest',
    'https://mydramalist.com/18452-goblin',
    'https://mydramalist.com/32925-hotel-del-luna',
    'https://mydramalist.com/24425-i-cannot-hug-you',
    'https://mydramalist.com/27173-hua-jai-sila',
    'https://mydramalist.com/52317-intense-love',
    'https://mydramalist.com/49865-psycho-but-it-s-okay',
    'https://mydramalist.com/692791-cai-bu-yao-he-lao-ban-tan-lian-ai',
    'https://mydramalist.com/39661-love-and-redemption'
    ]
```

## Web Scrapping

```python
# List to store the extrated data
all=[]

# Looping through each URL
for url in urls:
    page = requests.get(url,headers=headers)
    try:
        soup = BeautifulSoup(page.text, 'html.parser')
        # Extracting the drama Name and Year
        name_with_year=soup.find('h1',class_='film-title').text.strip()
        name=name_with_year[:-6].strip()
        year=name_with_year[-5:-1].strip()

        # Extracting show synopsis
        synopsis_container=soup.find('div',class_='show-synopsis')
        if synopsis_container:
          # Visible part
          synopsis_visible=synopsis_container.find('span').text.strip()
          # Hidden part
          synopsis_hidden_tag = synopsis_container.find('span', class_='show-more-hidden')
          synopsis_hidden = synopsis_hidden_tag.text.strip() if synopsis_hidden_tag else ""
          synopsis = synopsis_visible + " " + synopsis_hidden

        # Append the data into a list
        x=[name,year,synopsis]
        all.append(x)

    except AttributeError:
      print(f"Error while processing {url}")
    # Wait for a short time to avoid rate limiting
    time.sleep(1)
```

## Converting result into DataFrame

```python
Columns = ['Name','Year','Synopsis']
df = pd.DataFrame(all, columns=Columns)
for i in all:
  index=0
  df.loc[index]=i
  df.index-df.index+1
df=df.reset_index(drop=True)
df
```

## Saving the result in excel file.

```python
df.to_excel('MyDramaList.xlsx')
```
